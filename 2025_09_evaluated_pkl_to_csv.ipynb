```python
###
# This code extracts information from pkl files generated by YOLO performance evaluation and saves it to CSV format.
###
```

```python
import os, glob, shutil, h5py, cv2, random, tifffile, pprint 
import numpy as np, pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from joblib import Parallel, delayed
from PIL import Image, ImageFilter, ImageFile
from joblib import Parallel, delayed
from scipy import stats

ImageFile.LOAD_TRUNCATED_IMAGES = True
Image.MAX_IMAGE_PIXELS = None
%matplotlib inline
```

```python
d = '/PATH/TO/YOUR/OUTPUT/PKLs'
predicted_f = sorted(glob.glob(d + '/*.pkl'))
print(len(predicted_f))
```

```python
file_na = []
mean_precision = []
precision = []
mean_recall = []
recall = []
mean_ap50 = []
ap50 = []
mean_ap50_95 = []
ap50_95 = []
   
    
for i in predicted_f:
    with open(i, 'rb') as file:
        metric = pickle.load(file)
    file_na.append(os.path.basename(i))
    mean_precision.append(metric.mp)
    precision.append(metric.p)
    mean_recall.append(metric.mr)
    recall.append(metric.r)
    mean_ap50.append(metric.map50)
    ap50.append(metric.ap50)
    mean_ap50_95.append(metric.map)
    ap50_95.append(metric.ap)
```

```python
data = {
    'file_name': file_na,
    'mean_precision': mean_precision,
    'precision_0': [p[0] for p in precision],
    'precision_1': [p[1] for p in precision],
    'precision_2': [p[2] for p in precision],
    'precision_3': [p[3] for p in precision],
    'mean_recall': mean_recall,
    'recall_0': [p[0] for p in recall],
    'recall_1': [p[1] for p in recall],
    'recall_2': [p[2] for p in recall],
    'recall_3': [p[3] for p in recall],
    'mean_ap50': mean_ap50,
    'ap50_0':  [p[0] for p in ap50],
    'ap50_1':  [p[1] for p in ap50],
    'ap50_2':  [p[2] for p in ap50],
    'ap50_3':  [p[3] for p in ap50],
    'mean_ap50_95': mean_ap50_95,
    'ap50_95_0': [p[0] for p in ap50_95],
    'ap50_95_1': [p[1] for p in ap50_95],
    'ap50_95_2': [p[2] for p in ap50_95],
    'ap50_95_3': [p[3] for p in ap50_95]
}

df = pd.DataFrame(data)
```

```python
df.to_csv('SAVE_NAME.csv')
```
